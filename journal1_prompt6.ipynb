```python
import numpy as np
```

```python
def sin_func(x):
    return np.sin(x)
```

In the cell the code defines what the sin function is.

```python
def cos_func(x):
    return np.cos(x)
```

This cell then defines what the cos function is.

```python
x_values = np.linspace(0, 2*np.pi, 1000)

sin_values = sin_func(x_values)
cos_values = cos_func(x_values)
```

This cell will generate vauules for x in between 0 and 2pi where is then will calculate the sin and cos values with correlate.

```python
print(f"{'x':>10} | {'sin(x)':>10} | {'cos(x)':>10}")
print("-" * 35)

for i in range(10):
    print(f"{x_values[i]:10.5f} | {sin_values[i]:10.5f} | {cos_values[i]:10.5f}")
```

this final cell then prints the first ten values for x, sin(x), and cos(x) in a table.
